```python
import numpy as np
import matplotlib.pyplot as plt
data=np.random.random((13,16))
plt.imshow(data,cmap="magma")
plt.title("2-D Heatmap")
plt.colorbar()
plt.show()
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
data=np.random.randint(low=14,
                       high=100,
                       size=(10,10))
hm=sns.heatmap(data=data,annot=True)
plt.show()

```
